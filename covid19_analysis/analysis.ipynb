# Importing packages

```python
import pandas as pd
pd.set_option('display.float_format', '{:.2f}'.format)
```

# Importing dataset

```python
testing_link = "https://api.covid19india.org/csv/latest/statewise_tested_numbers_data.csv"
df = pd.read_csv(testing_link)
display(df.head())
display(df.describe(include='all'))
display(df.info())
```

```python
districts_link = 'https://api.covid19india.org/csv/latest/districts.csv'
df = pd.read_csv(districts_link)
display(df.head())
display(df.describe(include='all'))
display(df.info())
```
