# Scrapping financial data from screener.in

```python
from urllib.request import urlopen
import web_scrapping as ws
import pandas as pd
import numpy as np
```

```python
link = "https://www.screener.in/company/IOC/consolidated/"

page = urlopen(link)

html_bytes = page.read()
html = html_bytes.decode("utf-8")
```

## Quaterly Results

```python
def row_return(temp):
  rows = []
  temp1 = temp
  while temp1.find('<td class="text">') > 0:
    temp1 = temp1[temp1.find('<td class="text">'):]
    a = temp1[:temp1.find("</td>")]

    if a.find("&nbsp") != -1:
      a = a[a.find('this)">\n ')+8:a.find("&nbsp")]
    else:
      a = a[a.find('\n          \n')+12:]
      a = a[:a.find('\n          \n')]
    a = a.strip()
    rows.append(a)
    temp1 = temp1[temp1.find("</td>"):]
  return rows

def table_return(temp, title1, title2):
  headings_value, temp, rest = ws.getting_headers(temp, title1, title2)
  qr_index = row_return(temp)
  quarter_results = pd.DataFrame(columns = headings_value)

  for i in qr_index:
      values, temp = ws.func(temp, i, "td", headings_value)
      quarter_results.loc[i] = ws.string_to_float(values)

  return quarter_results
```

```python
quarter_results = table_return(html, "Quarterly Results", "Profit & Loss")
```
